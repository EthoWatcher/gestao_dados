```python
def get_qnt_quadros(df_marcar):
    def func_mapeia(linha):
        resultado = linha[0] - linha[1]
        return resultado


    saida_zipada_map = map(func_mapeia, zip(df_marcar["q_fim"], df_marcar["q_inicio"]))

    return np.array(list(saida_zipada_map))
```

```python
def get_list_aletorio_quadros_categoria(df_unidade, comportamento, qtd_quadros_selecionados = 15):
    def constroi_map(list_quadros_selecionados):
        def map_linha(quadro):
            r_quadro_linha = quadro in list_quadros_selecionados
            return r_quadro_linha

        return map_linha

#     comportamento = "Immobility"
    df_comportamento = df_unidade[(df_unidade["categoria"] == comportamento)]

    lis_quadro = df_comportamento["@f"].unique()
    qnt_quadros_list = len(lis_quadro)
    
    
    r_n_tem_quadros_suf = qnt_quadros_list < qtd_quadros_selecionados
    if(r_n_tem_quadros_suf):
        qtd_quadros_selecionados = qnt_quadros_list
#     print(lis_quadro)
#     print(len(lis_quadro))
    lis_quadros_rand_sele = np.random.choice(lis_quadro, qtd_quadros_selecionados, replace=False)

    serie = df_comportamento["@f"].map(constroi_map(lis_quadros_rand_sele))

    df_filtrado = df_comportamento[serie]
    return df_filtrado
```

```python
get_list_aletorio_quadros_categoria(df_unidade, "Immobility")
```

```python
def get_primeiro_ultimo_trecho(df_unidade, comportamento):
#     comportamento = "Immobility"
    df_comportamento = df_unidade[(df_unidade["categoria"] == comportamento)]

    tr_inicial = df_comportamento.iloc[0]
    tr_final = df_comportamento.iloc[len(df_comportamento["categoria"])-1]

    df_both = pd.DataFrame([tr_inicial, tr_final])
    return  df_both

get_primeiro_ultimo_trecho(df_unidade, "Immobility")
```

```python
def get_df_trechos(df_unidade, comportamentos):
    lis_df = []
    for comportamento in comportamentos:
        lis_df.append(get_list_aletorio_quadros_categoria(df_unidade, comportamento))
    return pd.concat(lis_df)
```

```python
def seleciona_quadros_sexo(df_femeas):
    dosagens = df_femeas["dosagem"].unique()

    lis_df = []
    for dosagem in dosagens:
        df_femeas_dosagens = df_femeas[(df_femeas["dosagem"] == dosagem)]
        unidades_da_dosagem = df_femeas_dosagens["unidade"].unique()
        for unidade in unidades_da_dosagem:
            df_quadros = get_df_trechos(df_femeas_dosagens[df_femeas_dosagens["unidade"] == unidade],["Immobility",'Swimming', 'Climbing','Headshaking', 'Diving'] )
            lis_df.append(df_quadros)

    saida = pd.concat(lis_df)
    return saida
# categorias = df["categoria"].unique()
# sexo =   df["sexo"].unique()
# unidade = df["unidade"].
# df_machos_veiculo = df[(df["dosagem"] == "veiculo") & (df["categoria"] == "Headshaking") & (df["unidade"] == "A1") ]
```
