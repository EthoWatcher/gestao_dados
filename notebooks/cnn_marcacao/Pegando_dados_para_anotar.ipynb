```python
def determina_quadro_original(linha):
    r_vd_igual= linha[0] == 0
    r_van_igual= linha[1] == 0
    r_quadro_repitido = r_vd_igual and r_van_igual
    return not r_quadro_repitido


saida_zipada_repetida = map(determina_quadro_original, zip(df["@Vd"], df["@Van"]))
```

```python
df_quadros_originais = df[ list(saida_zipada_repetida)]
```
