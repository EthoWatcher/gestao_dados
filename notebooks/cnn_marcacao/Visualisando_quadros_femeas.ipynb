```python
def get_image_str(vp_im):
#     print(vp_im)
    t = v.Tocador_video(PATH_ARQUIVOS, vp_im[0])
    img = t.get_quadro(vp_im[1])
    return img.decode("utf-8")
    
```

```python
df["img_str_b64"] = list(map(get_image_str, zip(df["id_video"],df["@f"])))
```

```python
def conv_to_string(l):
    l = l[1]
    s = f"""{{"id_experimento":"5f8f5503d3d64947e0da22e0",
              "nome":"localiza_animal",
              "marcacoes":{{}},
              "id_j":"{l["id_j"]}",
              "@f":"{l["@f"]}", 
              "id_video":"{l["id_video"]}", 
              "img_str_b64":"{l["img_str_b64"]}"
        }}"""
    return s
```

```python
s = list(map(conv_to_string,df.iterrows()))
```

```python
def saida_json(l_string_json):
    string = ""
    tama = len(l_string_json)
    for i, j_tx in enumerate(l_string_json):
        
        r_ultimo = i >= tama-1
        if r_ultimo:
            string = string + j_tx + "" 
        else:
            string = string + j_tx + ",\n" 
    s_saida= f"[{string}]"
    return s_saida

```

```python

with open('somefile.json', 'w') as the_file:
    the_file.write(saida_json(s))
```
