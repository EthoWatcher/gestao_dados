```python
import numpy as np
import pandas as pd 
```

```python
dados_experimento = pd.read_csv("machos.csv")
dados_experimento
```

```python
dados_veiculo = dados_experimento[dados_experimento["dosagem"] =="veiculo"]
dados_veiculo
```

```python
dados_flx = dados_experimento[dados_experimento["dosagem"] =="flx2.5mg"]
dados_flx
```

```python
duracao_imobilidade = dados_veiculo["d_e_Immobility_s"]
n = dados_veiculo["d_total_experimento"]
with Model() as imobilidade_model_veiculo:
#     probabilidade do animal adotar a postura de imobilidade vai de 0% até 100
    p_imobilidade = Uniform("p_imobilidade", lower=0, upper=1)
    imobilidade_dado_p = Binomial('imobilidade_dado_p', p=p_imobilidade, observed=duracao_imobilidade, n=n)
```

```python
from pymc3 import fit
```

```python
with imobilidade_model_veiculo:
    samples =  fit(random_seed=10000).sample(100000)
```

```python
from pymc3 import plot_posterior
```

```python
plot_posterior(samples, varnames=["p_imobilidade"], color="LightSeaGreen")
```

```python
duracao_imobilidade = dados_flx["d_e_Immobility_s"]
n = dados_flx["d_total_experimento"]
with Model() as imobilidade_model_flx:
#     probabilidade do animal adotar a postura de imobilidade vai de 0% até 100
    p_imobilidade = Uniform("p_imobilidade", lower=0, upper=1)
    imobilidade_dado_p = Binomial('imobilidade_dado_p', p=p_imobilidade, observed=duracao_imobilidade, n=n)
    samples_flx =  fit(random_seed=10000).sample(100000)
    
```

```python
plot_posterior(samples_flx, varnames=["p_imobilidade"], color="LightSeaGreen")
```
