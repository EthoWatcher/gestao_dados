# Abrindo o banco de dados original

```python
df_alterado = lendo_csv("banco_original_trabalho.csv")
df_alterado.head()
```

```python

def map_etografias(txt):
    x = re.split("Videos cegados/", txt)
    video = re.split(".avi", x[1])
    saida = f"{x[0]}Videos cegados/Etografias/{video[0]}.etoxml"
    return saida



def map_tking(txt):
    x = re.split("Videos cegados/", txt)
    video = re.split(".avi", x[1])
    saida = f"{x[0]}Videos cegados/Tracking/{video[0]}.tkin"
    return saida
```

```python
df_alterado["local_eto"] = df_alterado["local_avi"].apply(map_etografias) 
df_alterado["local_tkin"] = df_alterado["local_avi"].apply(map_tking) 
# df_alterado["local_tkin"][0]
df_alterado
```

```python
def get_dic_eto():
    ls_eto = df_alterado["local_eto"].unique()
    dic_saida ={}
    for path in ls_eto:
        dv = de.Descritores_Video(path,"Immobility")
        dic_saida[path] = dv.get_descritores_quadros_video()
    return dic_saida

def get_dic_tking():
    ls_traking = df_alterado["local_tkin"].unique()
    dic_saida ={}
    for path in ls_traking:
        dv = de.get_scala(de.abrindo_xml_root(path))
        dic_saida[path] = dv
    return dic_saida



```

```python
dic_eto = get_dic_eto()
div_trk = get_dic_tking()
print(dic_eto)
print(div_trk)
```
